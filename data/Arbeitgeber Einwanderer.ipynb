Importe

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

Welche Daten sind verf√ºgbar und Laden des Datensatzes

```python
col_list= ["us_economic_sector", "employer_name"]
visas_df = pd.read_csv("us_perm_visas.csv", usecols=col_list)
visas_df.head ()
```

Untersuchung der vorliegenden Daten im Datensatz


```python
fig, ax = plt.subplots()
fig.set_size_inches(12.7, 8.27)
sns.set_context(rc={"font.size":12,"axes.labelsize":13})
sns.countplot(x='employer_name', data=visas_df, palette = sns.cubehelix_palette(8, start=.5, rot=-.75), order=visas_df.employer_name.value_counts().iloc[:10].index)
plt.xticks(rotation=90)

#Iterating over elements in "employer_name" column and displaying counts above bars
for i, v in visas_df.employer_name.value_counts().head(10).reset_index().iterrows():
    ax.text(i,v.employer_name,v.unique()[1], horizontalalignment='center',verticalalignment='bottom')

ax.set(xlabel='Employer name', ylabel='Number of Visa applications')
```
